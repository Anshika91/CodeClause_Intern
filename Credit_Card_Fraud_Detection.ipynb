```python
# Importing all Dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset to pandas dataframe
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
# print rows of dataset
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# dataset info
credit_card_data.info()
```

```python
# check missing values in each column
credit_card_data.isnull().sum()
```

This dataset is highly unbalanced
it has two type of datasets__
0.0--> represents Normal or legit Transaction 
1.0--> represents Fradulent transaction

```python
# distribution of legit transaction and fradulent transaction
credit_card_data['Class'].value_counts()
```

```python
#separating the data for analysis
legit = credit_card_data[credit_card_data.Class==0]
fraud = credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measure of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare values for both transactions
credit_card_data.groupby('Class').mean()
```

UnderSampling 
Build a sample dataset containing similler distribution of normal and fradulent transactions

```python
# fradulent transaction are 183
legit_sample = legit.sample(n=183)
```

concatenating two dataframes

```python
# axis 0 means rows and axis 1 means column
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Spliting the data into feature and target

x= features and y= labels

```python
x = new_dataset.drop(columns='Class',axis=1)
y = new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

Split the data into training data and testing data

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape) 
```

Model Training

by Logistic Regression 

```python
model = LogisticRegression()
```

```python
#training the logistic regression model by traing data
model.fit(x_train, y_train)
```

Model Evaluation and accuracy score

```python
# accuracy on training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print('Training_data Accuracy:', training_data_accuracy)
```

```python
# accuracy on training data
x_test_prediction = model.predict(x_test)
testing_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print('Testing_data Accuracy:', testing_data_accuracy)
```
